```python
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.stattools import adfuller
```

```python
def generate_ar_process(lags,coefs,length):
    
    #cast coef to an np array
    coefs=np.array(coefs)
    
    #initial values
    series=[np.random.normal() for _ in range(lags)]
    
    for _ in range(length):
        #get previous values of the series, reserved
        prev_vals=series[-lags:][::-1]
        
        #get new value of the time series
        new_val=np.sum(np.array(prev_vals)*coefs)+np.random.normal()
        
        series.append(new_val)
        
    return np.array(series)
        
```

```python
def perform_adf_test(series):
    result=adfuller(series)
    print('ADF statistics: %f' % result[0])
    print('p-value: %f' % result[1])
```

```python
#if p<0.5 its is stationary else it is not stationary
```

## AR(1) Process

## stationary

```python
ar_1_process=generate_ar_process(1,[.5],100)
plt.figure(figsize=(10,4))
plt.plot(ar_1_process)
plt.title('sattionary AR(1) Process', fontsize=18)
```

```python
perform_adf_test(ar_1_process)
```

## Non-stationary

```python
ar_1_process=generate_ar_process(1,[1],100)
plt.figure(figsize=(10,4))
plt.plot(ar_1_process)
plt.title('stationary AR(1) Process', fontsize=18)
```

```python
perform_adf_test(ar_1_process)
```

## AR(2)

```python
ar_1_process=generate_ar_process(1,[.5,.3],100)
plt.figure(figsize=(10,4))
plt.plot(ar_1_process)
plt.title('stationary AR(2) Process', fontsize=18)
```

```python
perform_adf_test(ar_1_process)
```

```python
ar_1_process=generate_ar_process(1,[.7,.3],100)
plt.figure(figsize=(10,4))
plt.plot(ar_1_process)
plt.title('NON-stationary AR(1) Process', fontsize=18)
```

```python
perform_adf_test(ar_1_process)
```
